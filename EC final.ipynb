# Humidity

```python
import tensorflow as tf
from tensorflow.keras import layers
import pandas as pd
```

```python
dataset = pd.read_csv("Edge_Computing/dataset_humidity.csv")
dataset.head()
```

```python
#Dividing the dependent and independent variable


x = dataset.iloc[:,:-1].values   #all rows .. excluding last column 
y = dataset.iloc[:,-1:].values  #all rows .. taking last column      
```

```python
#print(dataset.iloc[1:2,0])
#print(dataset.iloc[1:2,2:3])
#print(x)
#print(y)
```

```python
#Creating Model

model = tf.keras.Sequential()
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1))


model.compile(optimizer='rmsprop', loss='mse', metrics=['mse'])

#
model.fit(x, y, epochs=100, batch_size=16)
```

```python
model.summary()
```

```python
model.save('Humidity_predictor_model')
```

```python
load_model = tf.keras.models.load_model('Humidity_predictor_model')

#converting the saved TF model to TF Lite and save it as a file with extension‘.tflite’.
converter = tf.lite.TFLiteConverter.from_keras_model(load_model)  

#Optimization to avoid errors
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()
```

```python
open("humidity_predictor.tflite", "wb").write(tflite_model)
```

# Temperature

```python
dataset = pd.read_csv("Edge_Computing/dataset_temperature.csv")
dataset.head()
```

```python
x = dataset.iloc[:,:-1].values   
y = dataset.iloc[:,-1:].values 
```

```python
#Creating Model

model = tf.keras.Sequential()
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1))


model.compile(optimizer='rmsprop', loss='mse', metrics=['mse'])

#
model.fit(x, y, epochs=100, batch_size=16)
```

```python
model.summary()
```

```python
model.save('Temperature_predictor_model')
```

```python
load_model = tf.keras.models.load_model('Temperature_predictor_model')

converter = tf.lite.TFLiteConverter.from_keras_model(load_model)  #Converts a TensorFlow model into TensorFlow Lite model.
converter.optimizations = [tf.lite.Optimize.DEFAULT]

tflite_model = converter.convert()
open("Temperature_predictor.tflite", "wb").write(tflite_model)
```
